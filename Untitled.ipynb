```python
import pandas as pd
import numpy as np
```

```python
# Load datasets
orders = pd.read_csv("olist_orders_dataset.csv")
customers = pd.read_csv("olist_customers_dataset.csv")
order_items = pd.read_csv("olist_order_items_dataset.csv")
products = pd.read_csv("olist_products_dataset.csv")
payments = pd.read_csv("olist_order_payments_dataset.csv")
reviews = pd.read_csv("olist_order_reviews_dataset.csv")
```

```python
# Merge tables
df = orders.merge(customers, on="customer_id", how="left") \
           .merge(order_items, on="order_id", how="left") \
           .merge(products, on="product_id", how="left") \
           .merge(payments, on="order_id", how="left") \
           .merge(reviews, on="order_id", how="left")
```

```python
#Removing duplicates
orders.drop_duplicates(inplace=True)
customers.drop_duplicates(inplace=True)
order_items.drop_duplicates(inplace=True)
products.drop_duplicates(inplace=True)
payments.drop_duplicates(inplace=True)
reviews.drop_duplicates(inplace=True)
```

```python
df["product_category_name"].fillna("unknown_category", inplace=True)
df["review_score"].fillna(0, inplace=True)
```

```python
df["order_purchase_timestamp"] = pd.to_datetime(df["order_purchase_timestamp"])
df["order_approved_at"] = pd.to_datetime(df["order_approved_at"])
df["order_delivered_customer_date"] = pd.to_datetime(df["order_delivered_customer_date"])
df["order_estimated_delivery_date"] = pd.to_datetime(df["order_estimated_delivery_date"])
```

```python
categorical_cols = ["customer_city", "customer_state", "product_category_name", "payment_type"]
for col in categorical_cols:
    df[col] = df[col].astype("category")
```

```python
# --- 1. First & Last Order Date per Customer ---
customer_orders = df.groupby("customer_id").agg(
    first_order_date=("order_purchase_timestamp", "min"),
    last_order_date=("order_purchase_timestamp", "max"),
    frequency=("order_id", "nunique")  # # of orders
).reset_index()
```

```python
# --- 2. Recency ---
max_date = df["order_purchase_timestamp"].max()
customer_orders["recency"] = (max_date - customer_orders["last_order_date"]).dt.days
```

```python
# --- 3. Monetary Value (Total Spend per Customer) ---
df["total_price"] = df["price"] + df["freight_value"]
monetary = df.groupby("customer_id")["total_price"].sum().reset_index()
monetary.rename(columns={"total_price": "monetary"}, inplace=True)
```

```python
# Merge monetary into customer_orders
customer_orders = customer_orders.merge(monetary, on="customer_id", how="left")
```

```python
# --- 4. Review Score Averages per Customer ---
review_avg = df.groupby("customer_id")["review_score"].mean().reset_index()
review_avg.rename(columns={"review_score": "avg_review_score"}, inplace=True)
customer_orders = customer_orders.merge(review_avg, on="customer_id", how="left")
```

```python
# --- 5. Add Customer Location Info (state & city) ---
customer_orders = customer_orders.merge(
    customers[["customer_id", "customer_city", "customer_state"]],
    on="customer_id",
    how="left")
```

```python
# --- 6. RFM Segmentation ---
# Score each RFM metric into quartiles (1=low, 4=high)
customer_orders["R_score"] = pd.qcut(customer_orders["recency"], 4, labels=[4,3,2,1])
customer_orders["F_score"] = pd.qcut(customer_orders["frequency"].rank(method="first"), 4, labels=[1,2,3,4])
customer_orders["M_score"] = pd.qcut(customer_orders["monetary"], 4, labels=[1,2,3,4])
```

```python
# Combine into RFM Score
customer_orders["RFM_score"] = (
    customer_orders["R_score"].astype(str) +
    customer_orders["F_score"].astype(str) +
    customer_orders["M_score"].astype(str))
```

```python
# --- 7. Customer Segmentation ---
def segment_customer(row):
    if row["R_score"] == "4" and row["F_score"] == "4" and row["M_score"] == "4":
        return "Champion"
    elif row["F_score"] in ["3","4"] and row["M_score"] in ["3","4"]:
        return "Loyal"
    elif row["R_score"] in ["3","4"] and row["F_score"] in ["1","2"]:
        return "At Risk"
    else:
        return "Others"
customer_orders["segment"] = customer_orders.apply(segment_customer, axis=1)
```
